## <b>Data Science | Pandas</b> musobaqasidaki masalalar yechimlari

```python
import pandas as pd

df = pd.read_excel('attempts.xlsx')
```

<b>A.</b> Pandas | Attempts | get_first_cpp

```python
df[df['lang'] == 'cpp'].iloc[0]
```

<b>B.</b> Pandas | Attempts | get_longest

```python
df[df['user'].str.len() == df['user'].str.len().max()]['user'].values
```

<b>C.</b> Pandas | Attempts | filter_attempts_by_username

```python
df[df['user'].str[0].isin(['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U'])]
```

<b>D.</b> Pandas | Attempts | get_even_id

```python
def digits_sum(n): return 0 if n == 0 else digits_sum(n // 10) + n % 10
```

```python
df[df['id'].apply(lambda value: digits_sum(value) % 2 == 0)]
```

<b>E.</b> Pandas | Attempts | get_most_unaccepted_count

```python
by_problem = df.groupby('problem_title')
by_problem['verdict'].apply(lambda s: 2*(s == 'Qabul qilindi').sum() < s.size).sum()
```

<b>F.</b> Pandas | Attempts | get_problems_count

```python
by_problem = df.groupby('problem_title')
by_problem['lang'].apply(lambda s: (s == 'cpp').sum() >= 2).sum()
```

<b>G.</b> Pandas | Attempts | get_today_attempts

```python
df[df['created'] >= '2022-07-22 00:00:00']  # 2022-07-22 o'rninga bugungi sanani yozish kerak
```

<b>H.</b> Pandas | Attempts | filter_attempts

```python
import re
df[df['problem_title'].apply(lambda title: re.search(r'[^a-zA-Z][^a-zA-Z]', title) is not None)]
```

<b>I.</b> Pandas | Attempts | get_users_solved

```python
by_solved_user = df[df['verdict'] == 'Qabul qilindi'].drop_duplicates(['user', 'problem_title']).groupby('user')
by_solved_user['id'].count().sort_values()
```

<b>J.</b> Pandas | Attempts | get_users_rating

```python
by_solved_user = df[df['verdict'] == 'Qabul qilindi'].drop_duplicates(['user', 'problem_title']).groupby('user')

balls = {
    'Boshlangich': 1,
    'Asoslar': 2,
    'Normal': 3,
    "O'rtacha": 4,
    "Ilg'or": 5,
    'Qiyin': 6,
}

by_solved_user['problem_difficulty'].apply(lambda s: s.apply(lambda value: balls[value]).sum()).sort_values()
```
